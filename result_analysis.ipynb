```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

# Visualize T5

```python
# read the summaries of T5 experiments
result_df = pd.read_csv('archive/wandb_export_2023-05-20T13 51 32.415+08 00.csv')
result_df
```

```python
# prepare the column of run_name
run_names = ['mt5-xl-{}-shot'.format(x) for x in [1, 2, 3, 5]] + ['mt5-xxl-{}-shot'.format(x) for x in [1, 2, 3, 5]]
run_columns = []
for name in run_names:
    for j in range(5):
        run_columns.append(name)
result_df['run_name'] = run_columns[::-1]
result_df = result_df.iloc[::-1]
result_df
```

```python
# plot the graph
fig, axes = plt.subplots(3, 1, figsize=(10, 7))

sb.barplot(data=result_df, x='run_name', y='rouge_1_fmeasure', ax=axes[0])
sb.barplot(data=result_df, x='run_name', y='rouge_2_fmeasure', ax=axes[1])
sb.barplot(data=result_df, x='run_name', y='rouge_L_fmeasure', ax=axes[2])

plt.savefig('mt5.jpg', dpi=500)
```
